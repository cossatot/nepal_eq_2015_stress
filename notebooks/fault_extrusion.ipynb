```python
import pandas as pd
import json
import halfspace.projections as hsp
import pyproj as pj
import numpy as np
from copy import deepcopy
from scipy.interpolate import griddata
from scipy.ndimage import map_coordinates
import halfspace.load as hs
import halfspace.sandbox as hbx
import matplotlib.pyplot as plt
import gdal
import pyproj as pj
```

```python
ht_file = '../data/fault_data/ht3d_wedge_utm45n.geojson'
```

```python
with open(ht_file) as f:
    ht = json.load(f)
```

```python
hte = deepcopy(ht)
```

```python
dem_f = '../data/dem/nepal_eq_utm45n_500m_clip.tif'
```

```python
dem_dataset = gdal.Open(dem_f)
```

```python
dem = dem_dataset.GetRasterBand(1).ReadAsArray()
```

```python
dem = np.flipud(dem)
```

```python
dem_transform = {'upper_left_x': dem_dataset.GetGeoTransform()[0],
                 'x_res_m' : dem_dataset.GetGeoTransform()[1],
                 'x_rotation': dem_dataset.GetGeoTransform()[2],
                 'upper_left_y': dem_dataset.GetGeoTransform()[3],
                 'y_rotation': dem_dataset.GetGeoTransform()[4],
                 'y_res_m': dem_dataset.GetGeoTransform()[5],
                 'n_cols': dem_dataset.RasterXSize,
                 'n_rows': dem_dataset.RasterYSize}

dem_transform['y_res_m'] *= -1 # correct for the upper-left origin thing

print(dem_transform)
```

```python
dem_transform['east_min'] = dem_transform['upper_left_x']

dem_transform['east_max'] = (dem_transform['x_res_m'] * dem_transform['n_cols']
                             + dem_transform['east_min'])

dem_transform['north_max'] = dem_transform['upper_left_y']

dem_transform['north_min'] = (-1 * dem_transform['y_res_m'] * dem_transform['n_rows']
                              + dem_transform['north_max'])
```

```python
e_idx = np.linspace(dem_transform['east_min'], dem_transform['east_max'], 
                    num=dem_transform['n_cols'])

n_idx = np.linspace(dem_transform['north_min'], dem_transform['north_max'],
                    num=dem_transform['n_rows'])
```

```python
plt.figure(figsize=(15,15))
plt.pcolormesh(e_idx, n_idx, dem)
plt.colorbar()
for ft in hte['features']:
    coords = np.asarray(ft['geometry']['coordinates'])
    plt.plot(coords[:,0], coords[:,1], 'r')
plt.axis('equal')
```

```python
ht_extrude = deepcopy(hte)

z_max = 15000

pt_depths = np.arange(0, z_max, step=1500)


for feat in ht_extrude['features']:
    coords = np.asarray(feat['geometry']['coordinates'])
    #print(feat['properties']['fault name'], coords.shape)
    
    feat['properties']['strike'] = hsp.get_strike_from_pts(lon_pts=coords[:,0],
                                                           lat_pts=coords[:,1])
    
    if feat['properties']['Name'] == 'Nyonno Ri Detachment':
        dip = 45
        
    elif feat['properties']['Name'] == 'Ama Drime Detachment':
        dip = 40
    else:
        dip = 55
    
    feat['properties']['dip'] = dip

    lon_z, lat_z, z_z = hsp.extrude_fault_trace(
                                            lon_pts=coords[:,0],
                                            lat_pts=coords[:,1],
                                            elev_pts=coords[:,2],
                                            depth_vec=pt_depths,
                                            strike=feat['properties']['strike'],
                                            dip=dip,
                                            h_coords = 'm',
                                            output_shape='vector')
    
    coords = np.hstack( (lon_z[:, np.newaxis], lat_z[:, np.newaxis], 
                         z_z[:, np.newaxis]) )
    feat['geometry']['coordinates'] = coords.tolist()
    feat['geometry']['type'] = 'MultiPoint'
```

```python
def df_from_dict(dic):
    df = pd.DataFrame(dic['geometry']['coordinates'],
                      columns=['east', 'north', 'depth'])
    df['point_index'] = df.index.values
    df['strike'] = dic['properties']['strike']
    df['dip'] = dic['properties']['dip']
    df['rake'] = -90.
    df['fault_name'] = dic['properties']['Name']
    df['htm_ind'] = dic['properties']['Id']
    
    return df
```

```python
fault_df = pd.concat([df_from_dict(feat) for feat in ht_extrude['features']],
                     ignore_index=True)
```
